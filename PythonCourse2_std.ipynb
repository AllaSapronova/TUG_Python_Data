```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

# Variance, Standatd Deviation

```python
data = [10,9,9.5,8.5,8,4]
```

```python
plt.plot(data, marker='o')
```

```python
mean = np.mean(data)
variance = np.var(data)
st_dev = np.std(data)

mean, variance, st_dev
```

```python
plt.plot(data, marker='o')
plt.hlines(mean, 0, len(data)-1, color = 'red', label='mean')

plt.hlines(mean-st_dev, 0, len(data)-1, linestyles='--', color='green', label='std')
plt.hlines(mean+st_dev, 0, len(data)-1, linestyles='--', color='green')

plt.legend()
plt.show()
```

# Confidence interval, True Mean

```python
x_bar = np.mean(data) #mean of data
s = np.std(data) # std of data
n = len(data) # number of samples
#z = 1.96 # for a 95% CI

z = 2.576 # for a 99% CI

lower = x_bar - (z * (s/math.sqrt(n)))
upper = x_bar + (z * (s/math.sqrt(n)))
med = np.median(data)
```

```python
plt.plot(data, marker='o')
plt.hlines(mean, 0, len(data)-1, color = 'red', label='mean')

plt.hlines(mean-st_dev, 0, len(data)-1, linestyles='--', color='green', label='std')
plt.hlines(mean+st_dev, 0, len(data)-1, linestyles='--', color='green')

plt.hlines(lower, 0, len(data)-1, linestyles='dashdot', color='orange', label='CI')
plt.hlines(upper, 0, len(data)-1, linestyles='dashdot', color='orange')

plt.hlines(med, 0, len(data)-1, linestyles='solid', color='blue', label='median')

title = ('CI for z ='+str(z))

plt.title(label=title)
plt.legend()
plt.show()
```

## median = 50th percentile
### n = (P/100) x N
#### N = number of values in the data set, 
#### P = percentile,
#### n = ordinal rank of a given value (with the values in the data set sorted from smallest to largest)

```python
np.percentile(data, 50)== med
```

```python
np.percentile(data, 50) , med
```

### n-th value in the ORDERED data set shows that P % of all data has _this_ or _lower_ value

```python
plt.plot(data, marker='o')
plt.hlines(np.percentile(data, 77), 0, len(data)-1, linestyles='--', color='black')
title = ('77th percentile of the dataset')

plt.title(label=title)
plt.show()
```
