### int_a/int_b = 0.1
### [int_a/int_b] * 2 =  [ int_a/int_b, int_a/int_b ] = [ 0.1 ,0. 1 ]
### sum ( [ int_a/int_b] * 2) = sum ( [ int_a/int_b, int_a/int_b ] ) = int_a/int_b + int_a/int_b = 0.1 + 0.1 = 0.2
 

```python
sum([(int_a/int_b)]*2)
```

```python
sum([(int_a/int_b)]*2) == 0.2
```

```python
sum([(int_a/int_b)] * 10) == 1
```

#### In the case of 1/10, the binary fraction is 3602879701896397 / 2 ** 55 which is close to but not exactly equal to the true value of 1/10.
