```python
import pandas as pd
import numpy as np
import random
```

```python
rocks = ['limestone','granit','marble','sandstone','obsidian']
r_types = ['sediment','ingeous','metamorphic','sediment','ingeous']
```

```python
data_len = 200
r = []
for i in range(data_len):
    r.append(random.randrange(0, len(rocks))) #randomly select one from rocks
```

```python
data = pd.DataFrame([rocks[i] for i in r], columns = ['rocks']) #make it a table
data['types'] = [r_types[i] for i in r] #add corresponding rock type
data['fracture'] = [random.randrange(0,5) for i in range(len(data))] #add fracture [0:5]
```

```python
#add some noise, disregard the warning!
nans = random.sample(range(len(data)), random.randrange(0,int(len(data)/2) ) )

for i in range(len(nans)):
    n = nans[i]
    data['fracture'].loc[n]= np.NaN
```
