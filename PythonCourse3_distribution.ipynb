```python
import numpy as np
import random
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

## uniform distribution

```python
from scipy.stats import uniform
```

```python
# random numbers from uniform distribution
n = 1000
start = 1
width = 20
dataU = uniform.rvs(size=n, loc = start, scale=width)
print(np.mean(dataU), np.std(dataU))
```

```python
plt.hist(dataU, color = 'blue', edgecolor = 'black', bins = 50) #try bins = 50
```

```python
ax = sns.distplot(dataU,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 1,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

## normal distribution

```python
from scipy.stats import norm
# generate random numbers from N(0,1)
dataN = norm.rvs(size=1000,loc=0,scale=1)
print(np.mean(dataN), np.std(dataN))
```

```python
plt.hist(dataN, color = 'blue', edgecolor = 'black')
```

```python
ax = sns.distplot(dataN,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

# Gamma Distribution

```python
from scipy.stats import gamma
dataG = gamma.rvs(a=3, size=1000) #try a=10, 1
print(np.mean(dataG), np.std(dataG))
```

```python
plt.hist(dataG, color = 'blue', edgecolor = 'black', bins = 50)
```

```python
ax = sns.distplot(dataG,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Gamma Distribution', ylabel='Frequency')
```

## Exponential Distribution

```python
from scipy.stats import expon
dataE = expon.rvs(scale=1,loc=0,size=1000) #try loc=10, try scale=10
print(np.mean(dataE), np.std(dataE))
```

```python
plt.hist(dataE, color = 'blue', edgecolor = 'black', bins = 50)
```

```python
ax = sns.distplot(dataE,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

## Binomial Distribution: a distribution where only two outcomes are possible, such as success or failure, and where the probability of success and failure is same for all the trials

```python
from scipy.stats import binom
dataB = binom.rvs(n=100,p=0.9,size=1000) # try p=0.5, p=0.8, n= number of trials
print(np.mean(dataB), np.std(dataB))
```

```python
plt.hist(dataB, color = 'blue', edgecolor = 'black')
```
